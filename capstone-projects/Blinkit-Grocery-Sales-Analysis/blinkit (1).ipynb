```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular'})
```

## **KPI'S**

```python
#total sales
total_sales = df['Sales'].sum()
print(f"Total Sales: {total_sales}")

#avg sales
avg_Sales = df['Sales'].mean()
print(f"Average sales : {avg_Sales}")

#No of items sold
no_of_items_sold = df['Sales'].count()
print(f"No.of itema = {no_of_items_sold}")

#avg of rating
avg_Rating = df['Rating'].mean()
print(f"Average Rating : {avg_Rating}")
```

## **CHART REQUIREMENT**

#### **Total Sales By Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat, labels= sales_by_fat.index, autopct = '%.1f%%', startangle = 90, colors = ['pink', 'purple'])
plt.title('Sales by fat content')
plt.axis('equal')
plt.show()
```

#### **Total Sales By Item Type**

```python
df.head()
```

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending = False)

plt.figure(figsize=(10,6))
bars = plt.bar(sales_by_type.index, sales_by_type.values, color = 'lightblue')

plt.xticks(rotation = -90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total_sales_by_type')

plt.show()
```

#### **Fat content by Outlet for Total Sales**

```python
df.head(1)
```

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular','Low Fat']]

ax= grouped.plot(kind = 'bar' , figsize = (8,5) , title = 'outley tier by item fat content', color = ['lightblue','purple'])
plt.xlabel('outlet location')
plt.ylabel('total sales')
plt.show()
```

```python
#### **Total Sales by outlet Establishment**
```

```python
df.head(1)
```

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize = (9,5))
plt.plot(sales_by_year.index , sales_by_year.values , marker = 'o' , linestyle = '-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('total sales')
plt.title(' outlet Establistment')
plt.tight_layout()
plt.show()
```

#### **Total Sales by outlet Size**

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize = (4,4))
plt.pie(sales_by_size ,labels =  sales_by_size.index , autopct = '%.1f%%', startangle = 90, colors = ['pink', 'purple','lightblue'])
plt.title('outlet size')
plt.tight_layout()
plt.show()
```

#### **Total Sales by outlet location**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending = False)

plt.figure(figsize = (8,3))
ax = sns.barplot( x = 'Sales', y = 'Outlet Location Type', data = sales_by_location, color = 'm')
plt.show()
```
