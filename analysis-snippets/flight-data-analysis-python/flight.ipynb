```python
# using bar graph airlines vs number of flights - horizontal bar graph
import matplotlib.pyplot as plt
data['airline'].value_counts(ascending = True).plot.barh(color = ['lightgreen', 'lightblue'])
plt.title('airline vs frequency')
plt.xlabel('frequencies')
plt.show()
```

bargraph departure and arrival time by frequecy

```python
data['departure_time'].value_counts(ascending = True) #showing departure time
```

```python
data['arrival_time'].value_counts(ascending = True) #showing arrival time
```

```python
#showing 2 subplots (departure and arrival time with their counts)

plt.figure(figsize = (16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index, data['departure_time'].value_counts().values, color = ('pink','purple') )
plt.title('departure time')
plt.xlabel('D.time')
plt.ylabel('D.freq')

plt.subplot(1,2,2)

plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values, color = ('Lightblue','purple'))
plt.title('arrival time')
plt.xlabel('A.time')
plt.ylabel('A.freq')
plt.show()
```

Bar graph representing source city and destination city

```python
data['source_city'].value_counts()
```

```python
data['destination_city'].value_counts()
```

```python
#subplot of source_city and destination_city with their counts
plt.figure(figsize=(16,4))

plt.subplot(1,2,1)

plt.barh(data['source_city'].value_counts().index, data['source_city'].value_counts().values, color = ('yellow','orange'))
plt.title('source_city')
plt.xlabel('no. of flights')
plt.ylabel('cities')

plt.subplot(1,2,2)
plt.barh(data['destination_city'].value_counts().index, data['destination_city'].value_counts().values, color=('darkblue','lightblue'))
plt.title('destination_city')
plt.xlabel('no. of flights')
plt.ylabel('cities')
plt.show()
          
```

price vs airline

```python
data.head()
```

```python
#grouping the airlines and checking their mean price

data.groupby('airline')['price'].mean()
```

```python
#areachart 
data.groupby('airline')['price'].mean().plot.area(color = 'lightcoral', alpha = 0.6)
plt.title('avg ticket price by airline')
plt.ylabel('avg.price')
plt.xlabel('airlines')
plt.show()

          
```

```python
#categorical plot 

sns.catplot( x='airline', y='price', kind = 'bar', palette = 'rocket', hue = 'class', data=data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    x='departure_time',
    y='price',
    kind='bar',
    data=data,
    palette='cool'   # âœ… lowercase and correct spelling
)

plt.title('Average Flight Price by Departure Time', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
